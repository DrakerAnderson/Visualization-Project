```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Read in CSV 

all_season_yards = pd.read_csv('Data/kickoff_returns_cleaned.csv')

all_season_yards.head()
```

```python
# List columns
all_season_yards.columns
```

```python
yds_by_team_data = all_season_yards.groupby('Team')['Yds'].describe()

yds_by_team_data
```

# Return Yards by Year by Team

```python
# Plot Data
df = pd.read_csv('Data/kickoff_returns_cleaned.csv')

# Create a scatter plot of yards by year filtered by team
plt.figure(figsize=(18, 12))
for team in df['Team'].unique():
    team_df = df[df['Team'] == team]
    plt.scatter(team_df['Year'], team_df['Yds'], label=team)

plt.xlabel('Year')
plt.xticks(rotation=90)  # Rotate x-axis tick labels by 90 degrees
plt.ylabel('Yards')
plt.title('Return Yards by Year by Team')
plt.xticks(range(df['Year'].min(), df['Year'].max()+1))  # Show all years on x-axis
plt.legend(title="Team", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Standard Deviation

```python
# Plot Standard Deviation with Min and Max Yards
teams = yds_by_team_data.index
means = yds_by_team_data['mean']
stds = yds_by_team_data['std']
mins = yds_by_team_data['min']
maxs = yds_by_team_data['max']

# Calculate the overall average yards
avg_yards = yds_by_team_data['mean'].mean()

plt.figure(figsize=(10, 6))
plt.errorbar(teams, means, yerr=stds, fmt='o', capsize=5)
plt.scatter(teams, mins, marker='v', label='min')
plt.scatter(teams, maxs, marker='^', label='max')

# Add a horizontal line to represent the average yards
plt.axhline(y=avg_yards, color='r', linestyle='--', label='avg')

plt.xlabel('Team')
plt.xticks(rotation=90)  # Rotate x-axis tick labels by 90 degrees
plt.ylabel('Yards')
plt.title('Total Return Yards by Team with Standard Deviation, Min, and Max')
plt.legend()
plt.show()
```

# Touchdowns by Year

```python
seasonal_stats_summary = pd.read_csv('Data\seasonal_stats_summary.csv')

df = pd.read_csv('Data\seasonal_stats_summary.csv')

# Group the data by year and touchdowns, and count the number of occurrences
touchdowns_by_year = df.groupby(['Year', 'Touchdowns']).size().reset_index(name='count')

# Pivot the data to create a table with years as columns and touchdowns as rows
touchdowns_by_year_pivot = touchdowns_by_year.pivot(index='Touchdowns', columns='Year', values='count')

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(touchdowns_by_year['Year'], touchdowns_by_year['Touchdowns'])
plt.title('Number of Touchdowns by Year')
plt.xlabel('Year')
plt.ylabel('Number of Return Touchdowns')
plt.show()
```

# Yards by Year by Team (1970-1980)

```python
# Ensure the 'Year' column is of datetime type
all_season_yards['Year'] = pd.to_datetime(all_season_yards['Year'], format='%Y')

# Filter the data to only include years between 1970 and 1980
all_season_yards_filtered = all_season_yards[(all_season_yards['Year'].dt.year >= 1970) & (all_season_yards['Year'].dt.year <= 1980)]

# Filter the data to only include yards
df_yards = all_season_yards_filtered[['Team', 'Year', 'Yds']]

# Convert the 'Year' column to a numeric value (year as an integer)
df_yards['Year'] = df_yards['Year'].dt.year

# Calculate the average yards for each year
average_yards_by_year = df_yards.groupby('Year')['Yds'].mean()

# Create a line plot with seaborn
plt.figure(figsize=(20,6))
sns.lineplot(data=df_yards, x='Year', y='Yds', hue='Team')

# Highlight the year 1974
plt.axvline(x=1974, color='r', linestyle='--', label='1974')

# Add a line for each year's average yards
plt.plot(average_yards_by_year.index, average_yards_by_year.values, color='black', linestyle=':', label='Average Yards by Year')

plt.title('Breakdown of Yards by Year by Team (1970-1980)')
plt.xlabel('Year')
plt.ylabel('Yards')
plt.legend(title="Team", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Yards by Year by Team All Years

```python
# Ensure the 'Year' column is of datetime type
all_season_yards['Year'] = pd.to_datetime(all_season_yards['Year'], format='%Y')

# Filter the data to only include years between 1970 and 2022
all_season_yards_filtered = all_season_yards[(all_season_yards['Year'].dt.year >= 1970) & (all_season_yards['Year'].dt.year <= 2022)]

# Filter the data to only include yards
df_yards = all_season_yards_filtered[['Team', 'Year', 'Yds']]

# Convert the 'Year' column to a numeric value (year as an integer)
df_yards['Year'] = df_yards['Year'].dt.year

# Calculate the average yards for each year
average_yards_by_year = df_yards.groupby('Year')['Yds'].mean()

# Create a line plot with seaborn
plt.figure(figsize=(20,6))
sns.lineplot(data=df_yards, x='Year', y='Yds', hue='Team')

# Highlight the year 1974 and 2011
plt.axvline(x=1974, color='r', linestyle='--', label='1974')
plt.axvline(x=2011, color='g', linestyle='--', label='2011')

# Add a line for each year's average yards
plt.plot(average_yards_by_year.index, average_yards_by_year.values, color='black', linestyle=':', label='Average Yards by Year')

# Set custom x-axis tick formatter to display all years
plt.xticks(range(1970, 2023))

plt.title('Breakdown of Yards by Year by Team (1970-2022)')
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('Yards')
plt.legend(title="Team", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
