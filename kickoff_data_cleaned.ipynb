```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Reference the file where the CSV is located
csv_path = Path("kickoff_returns_merged.csv")

# Import the data into a Pandas DataFrame
raw_df = pd.read_csv(csv_path, encoding="UTF-8")
raw_df.head()
```

```python
# drop uneccessary columns
cleaned_df = raw_df.drop(['20+', '40+'], axis=1)

cleaned_df.head()
```

```python
# rename columns
cleaned_df = cleaned_df.rename(columns={'Ret': 'Returns', 'KRet TD': 'Touchdowns', 'Lng': 'Longest_Return', 'FC': 'Fair_Catch', 'FUM': 'Fumbles'})
cleaned_df.head()
```

```python
# Replace 'T' with an empty string in 'Longest_Return'
cleaned_df['Longest_Return'] = cleaned_df['Longest_Return'].str.replace('T', '')

cleaned_df.head()
```
