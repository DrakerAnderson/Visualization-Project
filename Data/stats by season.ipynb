```python
#Import Dependancies
import pandas as pd
import numpy as np
from pathlib import Path

#Set the path for the CSV file
csv_path = Path("../Data/kickoff_returns_cleaned.csv")

#Read the CSV file into a DataFrame
df = pd.read_csv(csv_path)

#Display the first few rows of the DataFrame
df
```

```python
#Loop through data and get a sum of Yards by Year
kickoff_returns_by_year = df.groupby('Year')['Yds'].sum()

#Display the results
kickoff_returns_by_year.head(5)
```

```python
#loop through data and get a sum of touchdowns by Year
kickoff_touchdowns_by_year = df.groupby('Year')['Touchdowns'].sum()

#Display the results
kickoff_touchdowns_by_year.head(5)
```

```python
#loop through data and get a sum of returns by Year
kickoff_returns_count_by_year = df.groupby('Year')['Returns'].sum()	

#Display the results
kickoff_returns_count_by_year.head(5)
```

```python
#Merge the three DataFrames into one
kickoff_returns_summary = pd.DataFrame({
	'Yards': kickoff_returns_by_year,
	'Touchdowns': kickoff_touchdowns_by_year,
	'Returns': kickoff_returns_count_by_year
})

#Display the summary DataFrame
kickoff_returns_summary.head(5)
```

```python
#Save the data to a new CSV file
kickoff_returns_summary.to_csv("../Data/seasonal_stats_summary.csv")
```
